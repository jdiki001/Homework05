```python
import matplotlib.pyplot as plt
import scipy.integrate as integrate
from numpy import sin as sin
from numpy import cos as cos
import numpy as np

g = 9.8
dt = 0.01 
L1 = 1.0  
L2 = 1.0 
m1 = 1.0  
m2 = 1.0  
t_final = 10.0

def L(pos, t):
    #takes the derivative of each function
    #np.zeros_like sets L to same array as pos
    L = np.zeros_like(pos)
    
    #set list elements i.e. value for L at 0 = value for pos at 1
    L[0] = pos[1]

    shift1 = pos[2] - pos[0]
    shift2 = (m1+m2)*L1 - m2*L1*cos(shift1)*cos(shift1)
    
    L[1] = (((m2*L1*pos[1]*pos[1]*sin(shift1)*cos(shift1))) + (m2*g*sin(pos[2])*cos(shift1)) + ((m2*L2*pos[3]*pos[3]*sin(shift1))) - (((m1+m2)*g*sin(pos[0])))/shift2)

    L[2] = pos[3]

    shift3 = (L2/L1)*shift2
    L[3] = ((-m2*L2*pos[3]*pos[3]*sin(shift1)*cos(shift1)) + (m1+m2)*g*sin(pos[0])*cos(shift1) - ((m1+m2)*L1*pos[1]*pos[1]*sin(shift1)) - ((m1+m2)*g*sin(pos[2]))/shift3)

    return L

#time stlice
t = np.arange(0, t_final, dt)

theta1 = 43
theta2 = -23
omega1 = 0.0
omega2 = 0.0

#convert the angle from degrees to radians
pos = np.radians([theta1, omega1, theta2, omega2])

#integrate the ODE with variables L, pos, with respect to time
y = integrate.odeint(L, pos, t)

#position of pendulums
x1 = L1*sin(y[:,0])
y1 = -L1*cos(y[:,0])
x2 = x1+L2*sin(y[:,2])
y2 = y1+-L2*cos(y[:,2])

plt.plot(x1, y1, label='$Pendulum 1$')
plt.plot(x2, y2, label='$Pendulum 2$')
plt.title("Double Pendulum")
plt.xlabel('y-position')
plt.ylabel('x-position')
plt.legend()
plt.show()
```
